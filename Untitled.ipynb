```python
import os
import numpy as np
import matplotlib.pyplot as plt

path = r'E:\Google 雲端硬碟\Python\untitled\medical_ai\Final_project\Plan2\unet-master-2\data\membrane\train\label'
im = plt.imread(os.path.join(path,'4.BMP'))
plt.figure()
plt.imshow(im[:,:,0])
plt.show()

mask = np.zeros(np.shape(im))
mask[:,:,:][np.where(im[:,:,0] >= 200)] = 255
mask[:,:,:][np.where(im[:,:,1] >= 200)] = 0
plt.figure()
plt.imshow(mask)
plt.show()
```

```python
path = r'E:\Google 雲端硬碟\Python\untitled\medical_ai\Final_project\Plan2\unet-master-2\data\membrane\groundtrue'
save_path = r'E:\Google 雲端硬碟\Python\untitled\medical_ai\Final_project\Plan2\unet-master-2\data\membrane\test_mask'
for i in range(15):
    filename = str(i)+'.BMP'
    im = plt.imread(os.path.join(path,filename))
    
    mask = np.zeros(np.shape(im))
    mask[:,:,:][np.where(im[:,:,0] >= 200)] = 255
    mask[:,:,:][np.where(im[:,:,1] >= 200)] = 0

    
    plt.imsave(os.path.join(save_path,filename),mask/255)
```
